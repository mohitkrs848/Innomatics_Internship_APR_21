### Bivariate Analysis

```python
plt.figure(figsize=(20,10))
sns.relplot(x='DBNOs',y='kills',hue ='matchType',data =df)
plt.show()
```

```python
plt.figure(figsize=(18,10))
sns.lineplot(x='teamKills',y='kills',label='matchType',data=df)
plt.show()
```

Obs : Shows the relationship between user's kills and his overall team's Kills
